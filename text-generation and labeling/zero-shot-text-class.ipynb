```python
import pandas as pd
from transformers import pipeline
```

```python
# Load the file 
df = pd.read_csv('/kaggle/input/text-js-rep/text_gen_mmn.csv')
```

```python
classifier = pipeline("zero-shot-classification", model="facebook/bart-large-mnli")

# Define candidate labels
candidate_labels = ["satisfied", "not satisfied"]
```

```python
# Function to classify job satisfaction
def classify_satisfaction(text):
    result = classifier(text, candidate_labels)
    
    return result['labels'][0]

df['ZS_Label'] = df['Text'].apply(classify_satisfaction)
```

```python
df['ZS_Label'] = df['ZS_Label'].map({'satisfied': 1, 'not satisfied': 0})
```

```python
df2 = df[['Text', 'ZS_Label']]

# Save the resulting DataFrame to a CSV file
df2.to_csv('text_zero_shot.csv', index=False)
```
